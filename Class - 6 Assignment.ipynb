# One Hot Encoding

```python
df = pd.read_csv('Pizza Data.csv')
df.head()
```

```python
dummy = pd.get_dummies(df.Size) #Encoding Size coloumn
```

```python
dummy.head()
```

```python
dummy = pd.get_dummies(df.Size, drop_first=True)
dummy.head()
```

```python
df = df.drop('Size',axis=1)
```

```python
df
```

```python
df = pd.concat([df,dummy],axis=1)
df.head()
```

```python
df = pd.read_csv('Pizza Data.csv')
columns = ['Size']
for col in columns:
    one = pd.get_dummies(df[col],drop_first=True)
    df = pd.concat((df,one),axis=1).drop(col,axis=1)
```
