```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('supershops.csv')
df.head()
```

# MinMaxScaler / Normalization

```python
#MinMaxScaler / Normalization
from sklearn.preprocessing import MinMaxScaler
mmx = MinMaxScaler()
```

```python
df['Marketing Spend'] = mmx.fit_transform(df[['Marketing Spend']])
df.head()
```

```python
df['Administration'] = mmx.fit_transform(df[['Administration']])
df.head()
```

```python
df['Transport'] = mmx.fit_transform(df[['Transport']])
df.head()
```

```python
df['Profit'] = mmx.fit_transform(df[['Profit']])
df.head()
```
